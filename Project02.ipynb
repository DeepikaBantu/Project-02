```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data()
y_train_encoded=to_categorical(y_train)
y_test_encoded=to_categorical(y_test)
print('y_train_encoded shape:',y_train_encoded.shape)
print('y_test_encoded shape:',y_test_encoded.shape)
```

```python
y_train_encoded[0]
```

```python
import numpy as np
x_train_reshaped=np.reshape(x_train,(60000,784))
x_test_reshaped=np.reshape(x_test,(10000,784))
print('x_train_reshaped shape:',x_train_reshaped.shape)
print('x_test_reshaped shape:',x_test_reshaped.shape)
```

```python
print(set(x_train_reshaped[0]))
```

```python
x_mean=np.mean(x_train_reshaped)
x_std=np.std(x_train_reshaped)
epsilon=1e-10
x_train_norm=(x_train_reshaped-x_mean)/(x_std+epsilon)
x_test_norm=(x_test_reshaped-x_mean)/(x_std+epsilon)
```

```python
print(set(x_train_norm[0]))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model=Sequential([
    Dense(128,activation='relu',input_shape=(784,)),
    Dense(128,activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
model.compile(
    optimizer='sgd',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()
```

```python

model.fit(x_train_norm,y_train_encoded,epochs=3)
```

```python
x_mean=np.mean(x_train_reshaped)
x_std=np.std(x_train_reshaped)
epsilon=1e-10
x_train_norm=(x_train_reshaped-x_mean)/(x_std+epsilon)
x_test_norm=(x_test_reshaped-x_mean)/(x_std+epsilon)
_,accuracy=model.evaluate(x_test_norm,y_test_encoded)
print('Test set accuracy:',accuracy*100)
```

```python
preds=model.predict(x_test_norm)
print('Shape of preds:',preds.shape)
```

```python
plt.figure(figsize=(12,12))
start_index=0
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    pred=np.argmax(preds[start_index+i])
    gt=y_test[start_index+i]
    col='g'
    if pred!=gt:
        col='r'
    plt.xlabel('i={}, pred={}, gt={}'.format(start_index+i, pred, gt), color=col)
    plt.imshow(x_test[start_index+i],cmap='binary')
plt.show()
```

```python
plt.figure(figsize=(16, 16))  # Larger figure for 25 subplots

for i in range(25):
    plt.subplot(5, 5, i+1)  # 5x5 grid
    plt.plot(preds[i], 'b-', linewidth=1.5, marker='o', markersize=3)
    plt.title(f"Sample {i}", fontsize=9)
    plt.xlabel("Class", fontsize=7)
    plt.ylabel("Prob", fontsize=7)
    plt.grid(True, linestyle=':', alpha=0.4)

    # Highlight predicted class
    predicted_class = np.argmax(preds[i])
    true_class = y_test[i]
    pred_line = plt.axvline(x=predicted_class, color='r', linestyle='--', alpha=0.5, linewidth=1, label=f'Pred: {predicted_class}')

    # Highlight true class (if different from predicted)
    if predicted_class != true_class:
        true_line = plt.axvline(x=true_class, color='g', linestyle='-', alpha=0.5, linewidth=1, label=f'True: {true_class}')

    plt.xticks(range(len(preds[i])), fontsize=6)
    plt.yticks(fontsize=6)
    plt.ylim(0, 1)

    # Add legend (only if there's a mismatch to avoid clutter)
    if predicted_class != true_class:
        plt.legend(fontsize=6, loc='upper right')
    else:
        plt.legend([pred_line], [f'Pred: {predicted_class}'], fontsize=6, loc='upper right')

plt.tight_layout()  # Prevent overlapping
plt.show()
```
